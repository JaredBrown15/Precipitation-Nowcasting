```python
import numpy as np
import pandas as pd
np_load_old = np.load
np.load = lambda *a,**k: np_load_old(*a, allow_pickle=True, **k)
```

```python
load_path = 'D:\Final Dataset/threshold_114_nSTD_1/np_array_object/np_array.npy'
arr = np.load(load_path)
# arr
arr.shape
```

```python
first_frames = arr[:, 0:1, :]
first_frames = first_frames.reshape(first_frames.shape[0], first_frames.shape[2])

text_part = first_frames[:, :2]  # shape (17000, 2), strings
numeric_part = first_frames[:, 2:].astype(np.float64)  # shape (17000, 13), floats

# Combine into a DataFrame
df = pd.DataFrame(np.hstack((text_part, numeric_part)))

# Optional: add column names
df.columns = ['EventNumber', 'FileName'] + ['SemiMajorAxisLength', 
                'SemiMinorAxisLength',
                'Area',
                'AxisRatio',
                'EllipseAngle', 
                'COMX', 
                'COMY', 
                'Average',
                'Intensity',
                'VelocityX',
                'VelocityY',
                'Velocity Magnitude',
                'AngularVelocity']

df
# Save to CSV
```

```python
save_path = 'D:\Final Dataset/threshold_114_nSTD_1/np_array_object/first_frames.csv'
# np.savetxt(save_path, first_frames, delimiter=',')
df.to_csv(save_path, index=False)
```
