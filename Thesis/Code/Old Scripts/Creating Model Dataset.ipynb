Have to create files containing inputs and outputs.

Ahah jk, I just have to train on specified input length, produce output sequence of speficied length, then pull values from output

```python
import os
import pandas as pd
from pathlib import Path  
import time

path = "D:\Clustered Data/5"
dir_list = os.listdir(path)
```

```python
inputs = []
targets = []
inputTime = 60 # Length of time for input data sequence
leadTime = 60 # (Length of time between last data point and prediction) 
```

```python
input = int(inputTime/5)
lead = int(leadTime/5)
```

```python
0+(input+lead)
```

```python
start_time = time.time()

for file in dir_list:
    i = 0
    while i+input+lead<49:
        target_i = i+input+lead
        filepath = path + '/' + file
        df = pd.read_csv(filepath)    
        inputs.append(df.iloc[i:(i+input)])
        targets.append(df.iloc[target_i])
        i+=1
    # aggregate_df.append(line)
    # print('Adding first frame of' + file + 'csv')
```

```python
print("--- %s seconds ---" % (time.time() - start_time))
# inputs = pd.DataFrame(inputs)
targets = pd.DataFrame(targets)
```
