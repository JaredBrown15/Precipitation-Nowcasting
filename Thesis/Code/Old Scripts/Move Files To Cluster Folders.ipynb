# Move files based on clusters

```python
import os
from pathlib import Path  
import time
import pandas as pd
import shutil

data_path = "D:\Extracted Data"
dir_list = os.listdir(data_path)
catalog_path = Path("D:\Multivariate_Clustering.csv")
catalog = pd.read_csv(catalog_path)
catalog
```

```python
unclustered = []
for ind in catalog.index:
    eventID = catalog['EventID'][ind]
    cluster = catalog['Cluster'][ind]
    source = data_path + '/' + eventID + '_Params.csv'
    print(eventID)
    try:
        cluster = str(int(cluster))
        destination = "D:\Clustered Data/" + cluster + '/' + eventID + '_Params.csv'
        # print(source)
        # print(destination)
        shutil.copy(source, destination)
    except:
        print("No cluster detected for " + eventID)
        unclustered.append(eventID)


```
